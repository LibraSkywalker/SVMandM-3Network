## Problem 2

```python
import subprocess
from multiprocessing import Pool
from utility.core import *


if (__name__ == "__main__") :
    time0 = time.time()
    myPool = Pool(4)
    label,features = readNtranslate("./data/test.txt")
    
    for i in range(4) :
        folder = "GPUdata/Problem2/"
        file = "train"+str(i)+".txt"
        model_name = "tmp/model"+str(i)
        
        time1 = time.time()
        subprocess.run(args = ["./train",folder + file,model_name])
        time2 = time.time()
        print("train:",time2 - time1)
        
    for i in range(4) :
        model_name = "tmp/model"+str(i)
        
        myPool.apply_async(onlyPredict,(label,features,model_name))
        
    myPool.close()
    myPool.join()
    time3 = time.time()
    print("total:",time3 - time0)
    
    
```
